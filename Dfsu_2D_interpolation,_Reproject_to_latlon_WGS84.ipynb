**Interpolate dfsu to nc**

```python
!pip install mikeio xarray netcdf4
import os
import mikeio
#import xarray as xr
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
ds1 = mikeio.open("/content/drive/MyDrive/Colab Notebooks/WAVEtestBD.dfsu") 
ds1.items #liệt kê tên các biến
```

```python
ds = mikeio.read("/content/drive/MyDrive/Colab Notebooks/WAVEtestBD.dfsu",items=["x-comp. of wave vector","y-comp. of wave vector"]) #items là tên biến cần nội suy
ds
```

```python
da = ds[0]
da.plot();
```

```python
g = ds.geometry.get_overset_grid(dx=10000) #dx là kích thước pixel, vd 10000 m
g
```

```python
ds_grid = ds.interp_like(g) #nội suy dfsu thành raster
ds_grid
```

```python
ds_grid[0].plot();
```

```python
xr_ds = ds_grid.to_xarray()
xr_ds.to_netcdf("/content/drive/MyDrive/Colab Notebooks/wave vector.nc") # xuất sang Google Drive
```

**Reproject to WGS84**

```python
#import gdal
#!gdalinfo --version
#!pip install mikeio xarray netcdf4

!pip install rioxarray #cài đặt thư viện rioxarray
import rioxarray # for the extension to load
import xarray
import rasterio
```

```python
%matplotlib inline
xds = xarray.open_dataset("/content/drive/MyDrive/Colab Notebooks/wave vector.nc", decode_coords="all") #mở input
xds = xds.rio.write_crs("epsg:32648") #khai báo hệ tọa độ của input
xds
```

```python
xds_lonlat = xds.rio.reproject("epsg:4326") #khai báo hệ tọa độ của output
xds_lonlat
```

```python
xds_lonlat.to_netcdf("/content/drive/MyDrive/Colab Notebooks/wave vector_latlon.nc") #xuất output sang Drive
```
